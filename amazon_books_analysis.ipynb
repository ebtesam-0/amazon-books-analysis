```python
# الأعمدة التي تحتوي على قيم مفقودة وتريد تعبئتها
columns_to_fill = ['author', 'soldBy', 'publishedDate']

for column in columns_to_fill:  # كرر على كل عنصر في القائمة
    if df2[column].dtype == 'object':  # نفحص إذا العمود نصي
        mode_value = df2[column].mode()[0]  # نجيب أكثر قيمة متكررة
        df2[column] = df2[column].fillna(mode_value)  # نعبي القيم الفاضية بها
```

```python
# أول شي، نحول العمود 'publishedDate' لصيغة datetime
# عشان نقدر نسوي عمليات حسابية على التواريخ (زي الفرق بين تاريخين)
# errors='coerce' = أي شي غريب يتحول لـ NaN
df2['publishedDate'] = pd.to_datetime(df2['publishedDate'], errors='coerce')


from datetime import datetime  # استيراد مكتبة datetime للتعامل مع التواريخ
today = datetime.now()  # الحصول على تاريخ اليوم الحالي
df2['days_since_published'] = (today - df2['publishedDate']).dt.days  # حساب الفرق بين تاريخ اليوم و تاريخ النشر


# نطلع جدول فيه اسم الكتاب، تاريخ النشر، وعدد الأيام اللي مرت
books_days = df2[['title', 'publishedDate', 'days_since_published']]

# نعرض أول 10 كتب كمثال
display(books_days.head(10))
```

```python
# نبي نعرف أكثر المؤلفين عندهم كتب كثيرة في المكتبة
# value_counts() = يحسب عدد تكرار كل مؤلف في العمود 'author'
# head(15) = نجيب أول 15 مؤلف (الأكثر ظهوراً)
top_authors = df2['author'].value_counts().head(15)

print("كثر 15 مؤلفين نشاطًا حسب عدد الكتب:")

# نعرض النتائج مباشرة
# بدل print() نقدر نرجع top_authors لأنه DataFrame أو Series، بيطلع لنا جدول مرتب تلقائياً
top_authors
```

```python
# نحسب متوسط السعر لكل مؤلف
# groupby('author') = يجمع كل الكتب حسب المؤلف
# ['price'].mean() = يحسب المتوسط لكل مؤلف
# .round(2) = يقرّب الناتج لأقرب منزلتين عشريتين
avg_price_per_author = df2.groupby('author')['price'].mean().round(2)

# نعرض أول 10 مؤلفين مع متوسط سعر كتبهم
avg_price_per_author.head(15)
```

```python
# افتراضًا أن عمود 'reviews' يمثل النشاط
# نبي نحدد أكثر 10 مؤلفين لديهم أعلى عدد من التقييمات (النشاط)

#  هنا بنختار العمودين 'author' و 'reviews' من البيانات
#    لأننا نريد تحليل النشاط بناءً على التقييمات
top_active_customers = df2[['author', 'reviews']]  # اختيار الأعمدة المهمة فقط

#  بنجمع التقييمات لكل مؤلف باستخدام groupby()
top_active_customers = top_active_customers.groupby('author').sum()

# sort_values()   ترتيب البيانات في عمود أو أكثر حسب القيم
top_active_customers = top_active_customers.sort_values(by='reviews', ascending=False)
# ascending=True: يعني ترتيب تصاعدي، أي من الأقل إلى الأعلى.
# ascending=False: اذا اردت ترتيب تنازلي

#  نجيب أعلى 10 مؤلفين فقط (أكثر 10 مؤلفين لديهم أكبر عدد من التقييمات)
top_active_customers = top_active_customers.head(10)

#  عرض النتيجة (المؤلفين الأكثر نشاطًا حسب التقييمات)
top_active_customers
```

```python
#  اخترنا الأعمدة 'title' و 'reviews' فقط لأننا نبي نركز على أسماء الكتب وعدد التقييمات
#    اللي نعتبرها مبيعات المنتجات.
top_selling_products = df2[['title', 'reviews']]

#  استخدمنا sort_values() لترتيب المنتجات (الكتب) بناءً على عدد التقييمات من الأعلى إلى الأدنى
#    يعني بنعرض الكتب اللي حصلت على أكبر عدد من التقييمات.
top_selling_products = top_selling_products.sort_values(by='reviews', ascending=False)

#  استخدمنا head(10) عشان نجيب أول 10 كتب فقط اللي حصلوا على أعلى تقييمات (أكثر مبيعات).
top_selling_products = top_selling_products.head(10)

# عرض النتائج (أعلى 10 كتب حسب التقييمات)
top_selling_products
```

```python
# تحليل الاتجاهات الموسمية (عدد الكتب المنشورة حسب الشهر)


df2['month'] = df2['publishedDate'].dt.month  # استخراج الشهر من تاريخ النشر

# value_counts() تحسب عدد الكتب المنشورة في كل شهر
# sort_index() يرتب الأشهر من 1 إلى 12 عشان يكون التوزيع مرتب حسب الأشهر
books_per_month = df2['month'].value_counts().sort_index()

# رسم الرسم البياني الشهري
plt.figure(figsize=(12,6))
books_per_month.plot(kind='bar', color='red')
plt.title("Monthly Distribution of Books Published", fontsize=16)
plt.xlabel("Month", fontsize=12) # إضافة تسمية للمحور الأفقي (Month - الشهر)

plt.ylabel("Number of Books", fontsize=12) # إضافة تسمية للمحور العمودي (Number of Books - عدد الكتب)

plt.xticks(rotation=0)
# plt.xticks(): هذه الدالة تُستخدم لتحديد مواضع وعلامات المحور X (المحور الأفقي)
# rotation=0: تعني عدم تدوير علامات المحور X، أي أنها ستبقى في الوضع الطبيعي (أفقية).

plt.show()
```

```python
# هنا بنستخدم خاصية .dt.year لاستخراج السنة فقط من عمود التواريخ 'publishedDate'
df2['year'] = df2['publishedDate'].dt.year

# df2.dropna(): هذه الدالة تُستخدم لإزالة الصفوف أو الأعمدة التي تحتوي على قيم مفقودة (NaN).
# subset=['year']: يعني أنه سيتم فحص العمود year فقط. أي أنه سيتم إزالة الصفوف التي تحتوي على قيمة مفقودة في هذا العمود.
df2.dropna(subset=['year'], inplace=True)

# تحويل السنة إلى نوع بيانات صحيح (integer)
# بعد ما حذفنا القيم المفقودة، بنحول السنة من نوع float إلى int عشان نقدر نستخدمها في العمليات الحسابية
df2['year'] = df2['year'].astype(int)

# تصفية البيانات للسنوات من 1990 وما بعدها
# هنا بنفلتر البيانات عشان نجيب الكتب المنشورة فقط من سنة 1990 وما بعدdf_filtered_years = df2[df2['year'] >= 1990]
df_filtered_years = df2[df2['year'] >= 1990]


# بنستخدم value_counts() عشان نعرف كم كتاب تم نشره في كل سنة، وبنرتبهم من أقدم إلى أحدث باستخدام sort_index()
books_per_year = df_filtered_years['year'].value_counts().sort_index()

# Plot the bar chart
plt.figure(figsize=(12,6)) # نحدد حجم الرسم البياني
books_per_year.plot(kind='bar', color='red') # رسم الرسم البياني باستخدام bar chart

plt.title("Distribution of Books Published from 1990 Onwards", fontsize=16) # إضافة عنوان للرسم البياني
plt.xlabel("Year", fontsize=12) # إضافة تسمية للمحور الأفقي
plt.ylabel("Number of Books", fontsize=12) # إضافة تسمية للمحور العمودي

plt.xticks(rotation=90) # تدوير أسماء السنوات في المحور الأفقي بزاوية 90 درجة عشان تكون القراءة أسهل

plt.show()
```

```python
# Pie Chart
# Analyze most active customers (sellers)
# Get the top 10 sellers with the most books

# نستخدم value_counts() لحساب عدد الكتب عند كل بائع
top_sellers = df2['soldBy'].value_counts().head(5)

# رسم المخطط الدائري (Pie Chart)
 # عرض أفضل 10 بائعين

plt.figure(figsize=(12,6)) # تحديد حجم الرسم

top_sellers.plot(kind='pie',autopct='%1.1f%%')  # عرض النسب المئوية
plt.ylabel('')
plt.tight_layout()

plt.title("Top 5 Most Active Sellers", fontsize=16)
plt.show()
```

```python
# تحليل المنتجات الأكثر مبيعًا (الكتب)
# نطلع أكثر 10 كتب مبيعًا بناءً على تكرار العناوين

# نستخدم value_counts() لحساب عدد التكرارات لكل عنوان كتاب
top_books = df2['title'].value_counts().head(10)

# Plot the chart
# رسم الرسم البياني (Horizontal Bar Chart) لعرض أفضل 10 كتب مبيعًا

plt.figure(figsize=(12,6))
# تحديد حجم الرسم البياني

top_books.plot(kind='barh', color='gray')  # رسم الرسم البياني باستخدام barh (رسم بياني أفقي) عشان العناوين تكون أكثر وضوحًا

plt.title("Top 10 Most Frequent Books", fontsize=16)

plt.xlabel("Number of Copies", fontsize=12) # إضافة تسمية للمحور الأفقي  اللي يعبر عن عدد النسخ المباعة

plt.ylabel("Book Title", fontsize=12) # إضافة تسمية للمحور العمودي  اللي يعبر عن عنوان الكتاب

plt.gca().invert_yaxis()  # عكس ترتيب المحور العمودي بحيث يظهر الكتاب الأكثر مبيعًا في الأعلى

plt.show()
```
